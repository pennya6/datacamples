### Perceptron

```python
import numpy as np

def AND(x1,x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.7
    y=np.sum(w*x)+b
    if y>0:
        return 1
    else:
        return 0
    
def main():
    print("(0,0) = > ",AND(0,0))
    print("(0,1) = > ",AND(0,1))
    print("(1,0) = > ",AND(1,0))
    print("(1,1) = > ",AND(1,1))
    
main()
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def relu(z):
    if z>0:
        return z
    else:
        return 0

x=np.array([1.0,0.5])

w1=np.array([[0.1,0.3,0.5],[0.2,0.4,0.6]])
b1=np.array([0.1,0.2,0.3])

#가중치의 합
a1=np.dot(x,w1)+b1

#활성함수 적용
z1=sigmoid(a1)

print("가중치의 합 : ",a1)
print("활성함수 적용 : ",z1)
```

### Linear Regression

```python
import tensorflow as tf
import matplotlib.pyplot as plt

x=tf.constant([1,2,3,4,5],dtype=tf.float32)
y=tf.constant([1,2,3,4,5],dtype=tf.float32)

def run(w):
    hypothesis = w*x
    cost=tf.reduce_mean(tf.square(hypothesis-y))
    return cost.numpy()

w_val=[]
cost_val=[]

for i in range(-30,50):
    w=i*0.1
    c=run(w)
    
    if i%10==0:
        print('w = ',w,end='\t')
        print('c = ',c)
        
    w_val.append(w)
    cost_val.append(c)
    
plt.plot(w_val,cost_val)
plt.show()
```

### Linear Regression Tensorflow로 구현

#### 경사하강알고리즘 이용 -> 경사도에 따라 W를 변경시켜 cost값을 최소화하는 W를 구하는 알고리즘
#### 단 경사도는 미분값이다.

```python
# training data
x_data = tf.constant([[1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6, 2]], dtype=tf.float32)
y_data = tf.constant([[0], [0], [0], [1], [1], [1]], dtype=tf.float32)
W = tf.Variable(tf.random.normal([2, 1]), name='weight')
B = tf.Variable(tf.random.normal([1]), name='bias')

# define hypothesis
@tf.function
def Hypothesis(X):
    
    return tf.sigmoid(tf.matmul(X, W) + B)
```

```python
# define cost function
@tf.function
def loss(H, Y):
    cost = -tf.reduce_mean(Y * tf.math.log(H) + (1 - Y) * tf.math.log(1 - H))
    return cost
```

```python
# minimize the cost function
@tf.function
def train(X, Y, learning_rate=0.1):
    with tf.GradientTape() as tape:
        _loss = loss(Hypothesis(X), Y)
    _w, _b = tape.gradient(_loss, [W, B])
    W.assign_sub(learning_rate * _w)
    B.assign_sub(learning_rate * _b)
```

```python
for step in range(2001):
    train(x_data,y_data,learning_rate=0.01)
    _c=loss(Hypothesis(x_data),y_data)
    if step%20==0:
        print(f"{step}:{_c.numpy}{W.numpy()}{B.numpy()}")
print('\n final W=',W.numpy(),'b=',B.numpy())
```

```python
test_data=[2,4,1,5,3]
for data in test_data:
    y=data*W.numpy()+B.numpy()
    print("X=",data,"then Y=",y)
```

### Logisitic Regression

```python
x_data=tf.constant([[1,2],[2,3],[3,1],[4,3],[5,3],[6,2]],dtype=tf.float32)
y_data=tf.constant([[0],[0],[0],[1],[1],[1]],dtype=tf.float32)

W = tf.Variable(tf.random.normal([2,1]), name='weight')
B = tf.Variable(tf.random.normal([1]), name='bias')

@tf.function
def Hypothesis(X):
    return tf.sigmoid(tf.matmul(X,W)+B)

@tf.function
def loss(H, Y):
    cost=-tf.reduce_mean(Y*tf.math.log(H)+(1-Y)*tf.math.log(1-H))
    return cost

@tf.function
def train(X, Y, learning_rate=0.1):
    with tf.GradientTape() as tape:
        _loss = loss(Hypothesis(X), Y)
    _w, _b = tape.gradient(_loss, [W, B])
    W.assign_sub(learning_rate * _w)
    B.assign_sub(learning_rate * _b)
    
for step in range(10001):
    train(x_data,y_data,learning_rate=0.1)
    _c=loss(Hypothesis(x_data),y_data)
    if step%500==0:
        print(f"{step}:{_c.numpy()}")
```

```python
@tf.function
def test(H,Y):
    predicted=tf.cast(H>0.5,dtype=tf.float32)
    accuracy=tf.reduce_mean(tf.cast(tf.equal(predicted,Y),dtype=tf.float32))
    return predicted,accuracy

print("Accuracy...")
_h=Hypothesis(x_data)
_p,_a=test(_h,y_data)
print("Hypothesis=\n",_h.numpy())
print("Predicted=\n",_p.numpy())
print("\n Accuracy=",_a.numpy())
```
